```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("titanic-training-data.csv")
```

```python
df.head()
```

```python
df=df.drop(["Name"],axis=1)
df.head()
```

```python
## univariate
```

```python
sns.countplot(x="Age",data=df)
plt.show()
```

```python
df.describe()
```

```python
sns.boxplot(x="Age",y="Fare",data=df)
plt.show()
```

```python
sns.boxplot(x="Parch",y="Fare",data=df)
plt.show()
```

```python
sns.boxplot(x="Pclass",y="Fare",data=df)
plt.show()
```

```python
sns.boxplot(x="SibSp",data=df)
plt.show()
```

```python
plt.hist("Age",data=df)
plt.show()
```

```python
###Bivariate
```

```python
sns.countplot(x="Survived",hue="Pclass",data=df)
plt.show()
```

```python
sns.countplot(x="Survived",hue="Embarked",data=df)
plt.show()
```

```python
sns.relplot(x="Age",y="Fare",data=df)
```

```python
sns.relplot(x="Parch",y="Fare",data=df)

```

```python
sns.relplot(x="SibSp",y="Fare",data=df)
```
